```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# Step 1: Load the dataset
data = pd.read_csv('customer_purchase_data.csv')
```

```python
# Step 2: Display the first few rows and check column names
print(data.head())
print(data.columns)
```

```python
# Step 3: Use the correct target variable column name
target_column = 'PurchaseStatus'  # This is the correct column name
```

```python
# Step 4: Check for missing values
print(data.isnull().sum())
```

```python
# Step 5: Summary statistics
print(data.describe())
```

```python
# Step 6: Visualize target variable distribution
sns.countplot(x=target_column, data=data)
plt.show()
```

 This plot helps us to understand how the different outcomes of your target variable are distributed in your dataset.

```python
# Step 7: Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

The correlation heatmap shows which features in your data are related to each other. It helps you see which features go up and down together, so you can decide which ones are important for your model.

```python
# Step 8: Handle missing values (if any)
data.ffill(inplace=True)
```

```python
# Step 9: Label Encoding for categorical features
le = LabelEncoder()
```

```python
# Update with actual column names for categorical features
categorical_features = ['CategoricalFeature1', 'CategoricalFeature2']  # Replace with actual column names
for col in categorical_features:
    if col in data.columns:
        data[col] = le.fit_transform(data[col])
    else:
        print(f"Column '{col}' does not exist in the dataset")
```

```python
# Step 10: Feature Scaling
scaler = StandardScaler()
X = data.drop(target_column, axis=1)
X = scaler.fit_transform(X)
```

```python
# Step 11: Define target variable
y = data[target_column]
```

```python
# Step 12: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 13: Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
```

```python
# Step 14: Evaluation for Logistic Regression
print("Logistic Regression:")
print(classification_report(y_test, y_pred_logreg))
```

These lines of code create a Logistic Regression model, train it on our data, and then use it to make predictions about new, unseen data.

```python
# Step 15: Decision Tree
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred_dtree = dtree.predict(X_test)
```

```python
# Step 16: Evaluation for Decision Tree
print("Decision Tree:")
print(classification_report(y_test, y_pred_dtree))
```

It is used to predict results for new data.

```python
# Step 17: K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
```

```python
# Step 18: Evaluation for KNN
print("K-Nearest Neighbors:")
print(classification_report(y_test, y_pred_knn))
```

It is used to predict outcomes for new test data.

```python
# Step 19: Hyperparameter Tuning for Logistic Regression
param_grid = {'C': [0.1, 1, 10, 100], 'solver': ['liblinear']}
grid_logreg = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='accuracy')
grid_logreg.fit(X_train, y_train)

print(f"Best Parameters for Logistic Regression: {grid_logreg.best_params_}")
y_pred_logreg_tuned = grid_logreg.predict(X_test)
print("Tuned Logistic Regression:")
print(classification_report(y_test, y_pred_logreg_tuned))
```

```python
# Step 20: Hyperparameter Tuning for Decision Tree
param_grid = {'max_depth': [3, 5, 10, None], 'min_samples_split': [2, 5, 10]}
grid_dtree = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='accuracy')
grid_dtree.fit(X_train, y_train)

print(f"Best Parameters for Decision Tree: {grid_dtree.best_params_}")
y_pred_dtree_tuned = grid_dtree.predict(X_test)
print("Tuned Decision Tree:")
print(classification_report(y_test, y_pred_dtree_tuned))
```

```python
# Step 21: Hyperparameter Tuning for KNN
param_grid = {'n_neighbors': [3, 5, 7, 9], 'weights': ['uniform', 'distance']}
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_knn.fit(X_train, y_train)

print(f"Best Parameters for KNN: {grid_knn.best_params_}")
y_pred_knn_tuned = grid_knn.predict(X_test)
print("Tuned KNN:")
print(classification_report(y_test, y_pred_knn_tuned))
```
